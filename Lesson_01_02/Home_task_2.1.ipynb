1) Для любого y, принадлежащего отрезку [0,1], справедливо, что sgn(y)=1. sgn - функция определения знака числа.

Отрицание: Существует такое y, принадлежащее отрезку [0,1], что sgn(y) != 1


Доказательство: Если существует такое y, 0<y<1, что sgn(y) != 1, то y <= 0, что противоречит условию, а значит неверно.

2) Для любого натурального n, большего 2, существуют такие натуральные x, y и z, что x\*\*n = y\*\*n + z\*\*n

Отрицание: Существует такое натуральное n, большее 2, что для любых натуральных x, y и z, справедливо x\*\*n != y\*\*n + z\*\*n

Доказательство: Для любого натурального y и z мы можем такое x = pow(y\*\*n + z\*\*n, 1/n)

3) Для любого вещественного x существует такое вещественное Х, что Х > x

Отрицание:  Существует такое вещественное х, что для любого вещественного Х, X <= x

Доказательство: Если бы отрицание было бы истиной, это бы означало, что х - максимальное число. А максимального числа не существует.

4) Для любого комплексного x не существует такого комплексного y, что x > y или x < y

Отрицание: Для любого комплексного у, существует такое комплексное х, что x > y или x < y

Доказательство: Операция сравнения не определена для комплексных чисел.

Отрицание: Существует такое y принадлежащее отрезку [0, pi/2], что для любого ε > 0, sin(y) >= sin(y+ε)
